```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
sunspots=np.loadtxt('sunspots.txt',float)
nsunspots=np.array([sunspots[i][1] for i in range(3143)])
```

```python
plt.plot(nsunspots,'b-',linewidth=0.25,)
plt.xlabel('Date from 1749')
plt.ylabel('Observed Spots')
plt.title('Sunspots observed v/s Time')
```

```python
#only first 1000 datapoints
plt.plot(nsunspots[:1000],'b-',linewidth=0.5)
ns2=nsunspots[:1000]
```

```python
#running average Y=1/11*sig(-5,5)][y+m]
running_avg=[]
for i in range(1000):
    fact=0
    for y in range(-5,6):
        fact+=ns2[i]+y
    ans=1/11.0*fact
    running_avg.append(ans)
plt.plot(ns2,'b-', linewidth=0.5)
plt.plot(running_avg, 'r-', linewidth=0.5)
plt.show()
```
