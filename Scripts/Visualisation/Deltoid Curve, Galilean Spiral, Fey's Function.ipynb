```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib qt5
```

```python
#x=2cost+cos2t y=2sint -sin2t
theta=np.linspace(0,2*np.pi,1000)
x=2*np.cos(theta)+np.cos(2*theta)
y=2*np.sin(theta)+np.sin(2*theta)
plt.plot(x,y,'b')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# theta[0,10pi], r=theta^2, plot it
def cart(theta, r):
    x=r*np.cos(theta)
    y=r*np.sin(theta)
    return (x,y)
def func(theta):
    return theta**2
thetas=np.linspace(0,10*np.pi, 1000)
X=[]
Y=[]
for theta in thetas:
    x,y=cart(theta, func(theta))
    X.append(x)
    Y.append(y)
plt.plot(X,Y, 'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Galilean Spiral')
```

```python
#r= e^cost-2cos4t+sin^5(t/12) 0<t<24pi
thetas=np.linspace(0,24*np.pi,2400)
def cart(theta, r):
    x=r*np.cos(theta)
    y=r*np.sin(theta)
    return (x,y)
def func(theta):
    return np.exp(np.cos(theta))-2*np.cos(4*theta)+(np.sin(theta/12))**5
X=[]
Y=[]
for theta in thetas:
    x,y=cart(theta, func(theta))
    X.append(x)
    Y.append(y)
plt.plot(Y,X, 'b')
```
