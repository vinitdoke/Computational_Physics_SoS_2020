```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x,u):
    num=(1-u)**(-2)
    den=x**2+(u/(1-u))**2
    return num/den
def t(u):
    return u/(1-u)
a=0
b=1
N=1000
h=(b-a)/N
x=1
xval=[]
uval=np.linspace(0,1,N)
for u in uval:
    k1=h*f(x,u)
    k2=h*f(x+k1/2, u+h/2)
    k3=h*f(x+k2/2, u+h/2)
    k4=h*f(x+h, u+h)
    x+=1/6*(k1+2*k2+2*k3+k4)
    xval.append(x)
plt.plot(uval,xval,label="x v/s u")
tval=list(map(t,uval))


```

```python
plt.plot(tval,xval, label="x v/s t")
```
