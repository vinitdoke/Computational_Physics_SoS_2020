```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
NTl=1000
NPb=0
h=1
tau=3.053*60
p=1-2**(-h/tau)
tmax=1000
tpoints=np.arange(0.0,tmax,h)
Tlpoints=[]
Pbpoints=[]
for t in tpoints:
    Tlpoints.append(NTl)
    Pbpoints.append(NPb)
    decay=0
    for i in range(NTl):
        if  np.random.rand()<p:
            decay+=1
    NTl-=decay 
    NPb+=decay 

plt.plot(tpoints, Tlpoints, label="Thallium")
plt.plot(tpoints, Pbpoints, label="Lead")
plt.legend()
plt.title("Decay")
    

```
