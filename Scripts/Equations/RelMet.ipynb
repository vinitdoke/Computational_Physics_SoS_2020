```python
def f(x):
    return 5*np.exp(-x)+x-5
```

```python
err=10**(-6)
a=2
b=8
delta=abs(a-b)
while delta>err:
    c=(a+b)/2
    if f(c)*f(a)<0:
        b=c
    elif f(c)*f(b)<0:
        a=c
    delta=abs(a-b)
print((a+b)/2)
```

# Newton's Method
